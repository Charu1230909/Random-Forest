```python
import pandas as pd
```

```python
df=pd.read_csv("/content/cancer_data.csv")
print(f"\n Dataset Shape: {df.shape}")
df.head()
```

```python
X = df.drop(columns = 'diagnosis')
X
```

```python
from sklearn.model_selection import train_test_split
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
pred=dt.predict(X_test)
pred
```

```python
from sklearn.metrics import classification_report
cr=classification_report(pred,y_test)
print(cr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier (n_estimators=25)
rfc.fit(X_train,y_train)
```

```python
pred_rfc=rfc.predict(X_test)
pred_rfc
```

```python
rfc_cr=classification_report(pred_rfc,y_test)
print(rfc_cr)
```

```python
arg={
    'n_estimators':50,
    'max_features':"log2",
    'criterion':"entropy",
    'bootstrap':False,
    'max_depth':15,
    'min_samples_split':5,
    'min_samples_leaf':3,
}
from sklearn.ensemble import RandomForestClassifier
change_rf=RandomForestClassifier(**arg)
change_rf.fit(X_train,y_train)
```

```python
pred_change=change_rf.predict(X_test)
pred_change
```

```python
cf_change=classification_report(pred_change,y_test)
print(cf_change)
```
